```python
from pyspark.sql import SparkSession
import requests
from bs4 import BeautifulSoup

# Create a Spark session
spark = SparkSession.builder \
    .appName("ImageURLCounter") \
    .getOrCreate()

url = "https://newsroom.pinterest.com/en"
#url = "https://www.nationalgeographic.com/"

# Retrieving HTML content from a URL
def retrieve_html_content(url):
  try:
    #response=requests.get(url)
    return requests.get(url).text

  except Exception as e:
    print("Error retrieving html content from {url} : {str(e)}")

# Extracting image URLs from HTML content
def extract_img_urls(html_content):
  urls=[]
  beautifulsoup = BeautifulSoup(html_content, 'html.parser')
  img_tags = beautifulsoup.find_all('img')
  for tag in img_tags:
    srcs = tag.get('src')
    if srcs:
      urls.append(srcs)
  return urls

# Retrieving HTML content from the URL
html_content = retrieve_html_content(url)
#print(html_content)

# Checking whether html_content has fetched successfully
if html_content:

  # Extracting image URLs from HTML content
  img_urls = extract_img_urls(html_content)

  # Counting the number of image URLs
  img_urls_count = len(img_urls)

  print(f"Number of images in {url} : {img_urls_count}")

  # Displaying the image URLs
  c=1
  for imgurl in img_urls:
    print(f"Image {c} : {imgurl}")
    c+=1
```
