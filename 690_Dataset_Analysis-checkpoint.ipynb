```python
df = pd.DataFrame(data = data)
#sb.histplot(x = '')
x = data['Age']
y = data['Previous Purchases']
plt.scatter(x, y, label = 'ScatterPlot', color = 'blue', marker = 'o')
plt.xlabel('Age')
plt.ylabel('Previous Purchases')
plt.title('Age vs Previous Purchases')
plt.show()
```

```python
#seaborn.histplot
```

```python
# Define the age categories and labels
bins = [18, 30, 40, 50, 60, 70]
labels = ['18-30', '31-40', '41-50', '51-60', '61-70']

# Create a new column 'Age Group' with the categorized values
data['Age Group'] = pd.cut(data['Age'], bins = bins, labels = labels, right = False)

# Count the occurrences of each age group
age_group_counts = data['Age Group'].value_counts()

# Create a bar chart to visualize the distribution of age groups
plt.bar(age_group_counts.index, age_group_counts.values)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age vs Count')
plt.show()
```

```python
# Group the data by 'Category' and calculate the sum of 'Purchase Amount'
category_sum = df.groupby('Category')['Purchase Amount (USD)'].sum()

# Create a pie chart to visualize the sum of each category
plt.pie(category_sum, labels = category_sum.index, autopct = '%1.1f%%', startangle = 140)
plt.axis('equal')
plt.title('Total Purchase amount by category')
plt.show()
```

```python
payment_method_sum = df.groupby('Payment Method')['Customer ID'].sum()

plt.pie(payment_method_sum, labels = payment_method_sum.index, autopct = '%1.1f%%', startangle = 140, wedgeprops = dict(width = 0.4))
plt.axis('equal')
plt.title('Total count of Customers by Payment Method')
plt.show()
```

```python
# Filter the data for the 'Clothing' category
clothing_data = df[df['Category'] == 'Clothing']
#print(clothing_data.shape)
# Group the filtered data by 'Size' and sum the counts for each size
size_count = clothing_data['Size'].value_counts()
#print(size_counts)
size_count.plot(kind = 'barh', color = 'skyblue')

plt.xlabel('Count')
plt.ylabel('Size')
plt.title('Distribution of Clothing Sizes (Clothing Category)')
plt.show()

```

```python
# Define the criteria
subscription_yes = df['Subscription Status'] == 'Yes'
promo_used_yes = df['Promo Code Used'] == 'Yes'
previous_purchase_range = (df['Previous Purchases'] > 35) & (df['Previous Purchases'] <= 50)

# Apply the criteria to filter the DataFrame
filtered_df = df[subscription_yes & promo_used_yes & previous_purchase_range]
count = len(filtered_df)
print(count)
```

```python
# Group the data by "Shipping Type" and "Category" and count the occurrences
grouped_data = df.groupby(['Shipping Type', 'Category']).value_counts()

# Create the bar graph
grouped_data.plot(kind='bar', stacked=True, figsize=(8, 6))

plt.xlabel('Shipping Type')
plt.ylabel('Category')
plt.title('Shipping Type vs Category')
plt.show()
```

```python
# Group the data by "Category" and "Shipping Type" and count the occurrences
grouped_data = df.groupby(['Category', 'Shipping Type']).size().unstack()

# Create the bar graph
grouped_data.plot(kind='bar', figsize=(8, 6))

plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count of Categories in Each Shipping Type')
plt.legend(title='Shipping Type', loc='upper right')
plt.show()
```
