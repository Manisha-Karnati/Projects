```python
import pandas 
import seaborn
import matplotlib.pyplot as plt
import io
import requests
```

```python
url="https://roads.data.tfl.gov.uk/AccidentStats/Prod/2019-gla-data-extract-attendant.csv"
s=requests.get(url).content
data=pandas.read_csv(io.StringIO(s.decode('utf-8')))
data.head()
```

```python
d = data.drop(['Boro'],axis=1)
d.head()
```

```python
corelation = d.corr()
seaborn.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
d.dtypes
```

```python
d['Accident_date'] = pandas.to_datetime(d['Accident Date'])
d.head()
```

```python
d.head()
```

```python
df=pandas.DataFrame(data=d)
seaborn.histplot(x='Accident_date', data=df,hue="Day",multiple="dodge")
```

```python
df1=df[d['Accident Severity'].str.contains("1 FATAL")==True]
seaborn.histplot(x='Day', data=df1)
```

```python
df2=df[d['Accident Severity'].str.contains("2 SERIOUS")==True]
seaborn.histplot(x='Day', data=df2)
```

```python
df3=df[d['Accident Severity'].str.contains("3 SLIGHT")==True]
seaborn.histplot(x='Day', data=df3)
```
